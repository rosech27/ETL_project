```python
url = 'http://censusscope.org/2010Census/states.php?state=AZ&name=Arizona'


tables = pd.read_html(url)
tables
```

```python
total_pop = tables[0]
total_popbyrace= tables[1]
```

```python
total_popbyraceNEW = total_popbyrace.T
total_popbyraceNEW
```

```python
#remove row [0]rename header
total_popbyraceNEW.rename(columns={0:'Year',
                              1: 'White',
                              2: 'Black',
                              3: 'Hispanic',
                              4: 'American Indian',
                                5: 'Asian',
                               6: 'Hawaiian',
                               7:'Two or more',
                               8:'Other'}, inplace = True)
```

```python
total_popbyraceNEW.drop([0], inplace = True)
```

```python
total_popNEW = total_pop.T
total_popNEW
```

```python
#rename row[0] rename header
total_popNEW.rename(columns={0:'Year',
                              1: 'Population'}, inplace = True)
total_popNEW
```

```python
total_popNEW.drop([0], inplace =True)
total_popNEW
```

```python
total_popNEW.drop([1, 2, 3,4,5,6,7,8], inplace=True)
```

```python
# Merge two dataframes using an inner join
merged_table = pd.merge(total_popNEW, total_popbyraceNEW, on="Year")
merged_table
```

```python
merged_table.to_csv(r'ELT Project\PopulationRaceData.csv')
```
